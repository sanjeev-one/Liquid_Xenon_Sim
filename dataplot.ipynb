```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Define the column names
col_names = ['x', 'y', 'z', 'Px', 'Py', 'Pz', 't', 'PDGid', 'EventID', 'TrackID', 'ParentID', 'Weight']

# Read the file, skipping the first two rows
df = pd.read_csv('DetForward.txt', skiprows=3, delim_whitespace=True, names=col_names)
# Define the bin edges (from 0 to 100)
bin_edges = np.arange(0, 101, 1)  # Creates an array from 0 to 100 in steps of 1
df=df[df['PDGid']==-11]
# Plot a histogram of the 'Pz' column
df['Pz'].plot.hist(bins=bin_edges)
print(f"number of positrons: {df['Pz'].size}")
# Set the title and labels
plt.title('Histogram of Pz')
plt.xlabel('Pz')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
Edep = []
sum = 0
num_detectors = 11
num_detectors = num_detectors + 1
column_names = ["x", "y", "z", "Px", "Py", "Pz", "t", "PDGid", "EventID", "TrackID", "ParentID", "Weight", "Edep", "VisibleEdep", "Ntracks"]
for n in range(1,num_detectors):
    # Read the file, skipping the first two rows
    df = pd.read_csv(f'./disks/Det{n}.txt', skiprows=2, delim_whitespace=True, names=column_names)
    #print(df.columns)
    mean_value = df["Edep"].mean()
    sum_value = df["Edep"].sum()
    print(f"Sum value of Edep: {sum_value}")
    print("Mean value of Edep: ", mean_value)
    sum = sum + sum_value
    Edep.append(mean_value)
    
# Create a list of detector numbers for the x-axis
detectors = list(range(1, num_detectors))

plt.figure(figsize=(10, 6))
plt.plot(detectors, Edep, marker='o')

plt.title('Mean Energy Deposited in Each Detector')
plt.xlabel('Detector Number')
plt.ylabel('Mean Energy Deposited (Edep)')

plt.grid(True)
plt.show()

(print(f"Sum of Edep: {sum}"))
```
