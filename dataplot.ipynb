# lXe target sliced as disks

```python
#disks
Edep = []
sum = 0
num_detectors = 11
num_detectors = num_detectors + 1
column_names = ["x", "y", "z", "Px", "Py", "Pz", "t", "PDGid", "EventID", "TrackID", "ParentID", "Weight", "Edep", "VisibleEdep", "Ntracks"]
for n in range(1,num_detectors):
    # Read the file, skipping the first two rows
    df = pd.read_csv(f'./disks/Det{n}.txt', skiprows=2, delim_whitespace=True, names=column_names)
    #print(df.columns)
    mean_value = df["Edep"].mean()
    sum_value = df["Edep"].sum()
    print(f"Sum value of Edep: {sum_value}")
    print("Mean value of Edep: ", mean_value)
    sum = sum + sum_value
    Edep.append(mean_value)
    
# Create a list of detector numbers for the x-axis
detectors = list(range(1, num_detectors))

plt.figure(figsize=(10, 6))
plt.plot(detectors, Edep, marker='o')

plt.title('Mean Energy Deposited in Each Detector')
plt.xlabel('Detector Number')
plt.ylabel('Mean Energy Deposited (Edep)')

plt.grid(True)
plt.show()

(print(f"Sum of Edep: {sum}"))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the dimensions of your detector array
num_x = 11
num_y = 11
num_z = 21

# Initialize a 3D numpy array to hold the Edep sums
Edep_sums = np.zeros((num_x, num_y, num_z))

# Define column names
column_names = ["x", "y", "z", "Px", "Py", "Pz", "t", "PDGid", "EventID", "TrackID", "ParentID", "Weight", "Edep", "VisibleEdep", "Ntracks"]

# Loop over all detectors
for x in range(num_x):
    for y in range(num_y):
        for z in range(num_z):
            # Read the file, skipping the first two rows
            df = pd.read_csv(f'./square/Det{x}_{y}_{z}.txt', skiprows=2, delim_whitespace=True, names=column_names)
            sum_value = df["Edep"].sum()
            Edep_sums[x, y, z] = sum_value

# Now that we have the Edep sums, we can create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the x, y, and z coordinate arrays. We use numpy's indices to create three 3D arrays of coordinates.
x, y, z = np.indices((num_x, num_y, num_z))

# Convert the Edep sums to colors
color_values = Edep_sums / np.max(Edep_sums)
colors = plt.cm.viridis(color_values)

# Plot the scatter plot
ax.scatter(x.ravel(), y.ravel(), z.ravel(), c=colors.reshape(-1, 4), marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Energy Deposited in Each Cube')
plt.show()
```

### Peak Energy Deposition Density

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the dimensions of your detector array
num_x = 11
num_y = 11
num_z = 21

# Define the size of your cubes in cm
size_x = 1  # 10mm
size_y = 1  # 10mm
size_z = 1  # 10mm

# Define the density of LXE in g/cm^3
density = 2.953

# Calculate the mass of each cube in g
m_cube = density * size_x * size_y * size_z

# Initialize a 3D numpy array to hold the Edep sums
Edep_sums = np.zeros((num_x, num_y, num_z))

# Define the electron charge in C
e_charge = 1.60217663e-19

# Define column names
column_names = ["x", "y", "z", "Px", "Py", "Pz", "t", "PDGid", "EventID", "TrackID", "ParentID", "Weight", "Edep", "VisibleEdep", "Ntracks"]

# Loop over all detectors
for x in range(num_x):
    for y in range(num_y):
        for z in range(num_z):
            # Read the file, skipping the first two rows
            df = pd.read_csv(f'./square/Det{x}_{y}_{z}.txt', skiprows=2, delim_whitespace=True, names=column_names)
            sum_value = df["Edep"].sum() * 1E6 * e_charge / m_cube  # convert MeV to J/g
            Edep_sums[x, y, z] = sum_value

# Now that we have the Edep sums, we can create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the x, y, and z coordinate arrays. We use numpy's indices to create three 3D arrays of coordinates.
x, y, z = np.indices((num_x, num_y, num_z))

# Convert the Edep sums to colors
color_values = Edep_sums / np.max(Edep_sums)
colors = plt.cm.viridis(color_values)

# Plot the scatter plot
ax.scatter(x.ravel(), y.ravel(), z.ravel(), c=colors.reshape(-1, 4), marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Peak Energy Deposition Density in Each Cube - 200mm target')
plt.show()
plt.title('Peak Energy Deposition Density in Each Cube slice on the x-axis - 200mm target')
plt.legend('topright', 'Energy Deposition Density J/g')
plt.imshow(Edep_sums[4], cmap='viridis', origin='lower')

```

```python
import matplotlib.cm as cm
import matplotlib.colorbar as colorbar

fig, ax = plt.subplots(figsize=(10,8))

# Create a colormap
cmap = cm.viridis

# Normalize the color map based on your data's range
norm = plt.Normalize(vmin=np.min(Edep_sums[4]), vmax=np.max(Edep_sums[4]))

# Create an image using imshow and apply the colormap and normalization
img = ax.imshow(Edep_sums[4], cmap=cmap, norm=norm, origin='lower')

# Create a colorbar based on your image
cbar = fig.colorbar(img, ax=ax)
cbar.set_label('Energy Deposition Density (J/g)', rotation=270, labelpad=20)

plt.title('Peak Energy Deposition Density in Each Cube slice on the x-axis - 200mm target')
plt.show()
```
